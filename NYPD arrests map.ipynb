# NYPD arrests map

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n = pd.read_csv("NYPD_Arrests_Data__Historic_.csv")
```

```python
n.head()
```

```python
set(n['PERP_RACE'])
```

```python
n['colors'] = n['PERP_RACE'].map({'BLACK HISPANIC':'r','BLACK':'r',
                                  'WHITE HISPANIC':'b','WHITE':'b',
                                  'ASIAN/PAC.ISL':'g',
                                  'UNKNOWN':'w',
                                  'AMER IND':'w'})
```

```python
fig = plt.figure(figsize=(20,16))
ax = fig.add_subplot(111)

n.plot.scatter(x='Longitude',y='Latitude',s=1,alpha=.1,ax=ax,c=n['colors'])
ax.axis('off')
```
